```python
# Import necessary library
import requests  # For making API requests
import json      # For parsing JSON data
import pandas as pd  # For handling tabular data
```

```python
# API URL
#url = "https://api.sunrisesunset.io/json?lat=38.907192&lng=-77.036873&timezone=UTC&date=today"
url = "https://api.sunrisesunset.io/json?lat=-6.92222&lng=107.60694&date_start=2024-01-01&date_end=2024-12-31"
```

```python
# Fetch data from the API
try:
    response = requests.get(url)
    if response.status_code == 200:
        # Parse JSON data
        data = response.json()
        print("Data fetched successfully.")
    else:
        print(f"Error: {response.status_code}, {response.text}")
except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
```

```python
print(data)
```

```python
# Extract the 'results' key
results = data['results']

# Convert the extracted data to a Pandas DataFrame
df = pd.DataFrame(results)
```
